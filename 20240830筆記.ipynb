### 綜合練習 02
請將上面的程式碼打包成一個函式，輸入一個整數n，劃出高度為n的金字塔

````python
def add(a,b):
  print(f"{a}+{b}={a+b}")

def sub(a,b):
  print(f"{a}-{b}={a-b}")

def mul(a,b):
  print(f"{a}x{b}={a*b}")

def div(a,b):
  print(f"{a}/{b}={a/b}")
````

## 綜合練習 03

使用例外處理，修正萌典API的應用可能產生的錯誤

````python
import requests

def lookup(word):
  url="https://www.moedict.tw/uni/"
  data=requests.get(url + word).json()
  try:
    return data['heteronyms'][0]['definitions'][0]['def']
  except Exception as e:
    return '查無資料'
````

````python
lookup('apple')
````

### 計算1999到2024年，每一年的投資報酬率

````python
total = 1
for i in range(1999,2025):
  ret = df.loc[str(i)].Close.iloc[-1]/df.loc[str(i)].Close.iloc[0]
  total *= ret
  print(i,'==>',ret)
  total
````

### 下載股價歷史資料

````python
for i in range(12):
  ret = df.loc[f'2024-{i+1:02}'].Close.iloc[-1]/df.loc[f'2024-{i+1:02}'].Close.iloc[0]
  total *= ret
  print(f'2024-{i+1:02}','==>',ret)
````
